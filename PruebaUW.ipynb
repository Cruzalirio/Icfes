2. Hacer una tabla dinámica agrupando por curso los resultados por área. 

```python
Prueba.groupby("Grado").count()
```

Cómo todos son de grado 5, sólo se agrupa por curso e institución. Calculamos la media dentro de cada curso e institución

```python
Medias_curso = Prueba[['Institución',  'Curso', 'MATEMÁTICAS', 
                       'CIENCIAS SOCIALES', 'CIENCIAS NATURALES',
       'LENGUAJE']].groupby(["Institución", "Curso"], as_index=False).mean()
Medias_curso
```

Para saber los mejores resultados por area, usaré el promedio de los estudiantes como guía.


```python
Medias_curso.sort_values(by="MATEMÁTICAS", ascending=False)
## En matematicas es la IED JUAN JOSE RONDON
```

```python
Medias_curso.sort_values(by="CIENCIAS SOCIALES", ascending=False)
## En ciencias sociales es la IED DE CARRIZAL
```

```python
Medias_curso.sort_values(by="CIENCIAS NATURALES", ascending=False)
## En ciencias sociales es la IED DE CARRIZAL
```
